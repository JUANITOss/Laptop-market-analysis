```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set_theme()
```

```python
df = pd.read_csv('./data/bike_sales_100k.csv')

df.head()
```

```python
df.describe()
```

```python
cat_x1 = df[['Store_Location', 'Payment_Method', 'Customer_Gender']]

x1 = pd.get_dummies(cat_x1, drop_first=True).astype(int)
x1['Customer_Age'] = df['Customer_Age']
x1['Quantity'] = df['Quantity']

x1.head()
```

```python
y = df['Price']

y.head()
```

```python
x = sm.add_constant(x1)

results = sm.OLS(y, x).fit()

results.summary()
```
